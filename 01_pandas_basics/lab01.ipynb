## 4.1

```python
import pandas as pd
import numpy as np
import pickle
```

```python
df = pd.read_csv('lab1_ex01.csv')
json_array = []

type_dict = {"float64": "float", "int64": "int", "object": "other", "bool": "boolean"}

for column in df.columns:
    temp = df.loc[:,column]
    json_array.append([column, temp.isna().sum() / len(temp), type_dict[str(temp.dtype)]])


new_df = pd.DataFrame(json_array, columns=['name', 'missing', 'type'])
new_df.to_json('ex01_fields.json', orient='records')
df
```

## 4.3

```python
import re

replace_dict = {}
for column in df.columns:
    column_copy = re.sub(r'[^a-zA-Z0-9_ ]', '', column)
    column_copy = column_copy.lower()
    column_copy = column_copy.replace(" ", "_")
    replace_dict[column] = column_copy

df = df.rename(columns=replace_dict)
df.to_csv('ex03_columns.csv', index=False)
df
```
